Use la tabla hurto a personas por comuna/corregimiento del Informe Semanal de Hurtos en
Medellín. Para cada comuna 𝑖, defina su indicador estandarizado de la forma.

H_i_gorro=hi-h_gorro/Sh

hi_=cantidad de hurtos

Sh=desviacion estandar de hurtos por comuna/corregimiento



```python
import numpy as np
hurtos = [
    18,   # 01 Popular
    11,   # 02 Santa Cruz
    36,   # 03 Manrique
    79,   # 04 Aranjuez
    87,   # 05 Castilla
    33,   # 06 Doce de Octubre
    74,   # 07 Robledo
    29,   # 08 Villa Hermosa
    88,   # 09 Buenos Aires
    456,  # 10 La Candelaria
    276,  # 11 Laureles Estadio
    42,   # 12 La América
    35,   # 13 San Javier
    252,  # 14 El Poblado
    72,   # 15 Guayabal
    101,  # 16 Belén
    1,    # 50 San Sebastián de Palmitas
    25,   # 60 San Cristóbal
    15,   # 70 Altavista
    11,   # 80 San Antonio de Prado
    15    # 90 Santa Elena
]
hi=np.array(hurtos)
h_gorro=np.mean(hi)
Sh=np.std(hi)
H_i_gorro=(hi-h_gorro)/Sh
print("Promedio de hurtos por comuna/corregimiento: ",h_gorro)
print(f"Indicador estandarizado de hurtos por comuna/corregimiento: {H_i_gorro}")

print(len(A))
```

```python
# Crear matriz de pesos W como array de NumPy
W = np.zeros((len(A), len(A)), dtype=np.float64)
sigma = 1

for i in range(len(A)):
    for j in range(len(A)):
        if A[i,j] == 1: 
            diferencia = H_i_gorro[i] - H_i_gorro[j]
            # Fórmula correcta: exp(-((h_i - h_j)^2) / (2 * sigma^2))
            W[i,j] = np.exp(-(diferencia**2) / (2 * sigma**2))
        else:
            W[i,j] = 0

print(f"Número de aristas con peso no cero: {np.count_nonzero(W)}")
print("\nMatriz de pesos W:")
print(W)
```

```python

def jaccard_distance(A,B):
    return 1-len(A.intersection(B))/len(A.union(B))

distances = []
i=0
for _, row in df.iterrows():
    i+=1
    if row["title"] == "The Matrix Revolutions":
        continue
    TituloPeliculaActual = row["title"]
    GenerosPeliculaActual = row["genres"]
    distance = jaccard_distance(Matix_Generos, GenerosPeliculaActual)
    distances.append((TituloPeliculaActual, distance, GenerosPeliculaActual))
    distances.sort(key=lambda x: (x[1], x[0]))
top10 = distances[:10]
top=[]
for titulo, distancia, generos in top10:
    print(f"{titulo}: {distancia:.3f}")

for distance in distances:
    if distance[1] == 0.0:
        top.append(distance[0])

print(f"Cantidad de Peliculas con 0: {len(top)}")
```

2.2 Similaridad por Sinopsis

-  Use el campo overview como texto
-  defina la función levenshtein_distance(s1, s2) como el mínimo número de ediciones (insertar, borrar, sustituir) para transformar s1 en s2.



```python
import numpy as np

def levenshtein_distance(s1, s2):
    rows = len(s1) + 1
    cols = len(s2) + 1
    
    dist_matrix = np.zeros((rows, cols), dtype=int)

    # Inicializar primera fila y columna
    for i in range(rows):
        dist_matrix[i, 0] = i
    for j in range(cols):
        dist_matrix[0, j] = j

    # Llenar matriz
    for i in range(1, rows):
        for j in range(1, cols):

            if s1[i-1] == s2[j-1]:
                cost = 0
            else:
                cost = 1

            dist_matrix[i, j] = min(
                dist_matrix[i-1, j] + 1,      # borrar
                dist_matrix[i, j-1] + 1,      # insertar
                dist_matrix[i-1, j-1] + cost  # sustituir
            )

    return dist_matrix[rows-1, cols-1]
```

- Calcule 𝑑𝐿 entre el overview de “The Matrix Revolutions” y el de las demás películas.
- Reporte el top 10 de las películas con menor distancia

```python
Matrix_Overview=str(df[df["title"]=="The Matrix Revolutions"].iloc[0]["overview"])[:150]


# Acotar a 300 filas incluyendo la película de referencia (fila 125)
ref_iloc = 125
start = max(0, ref_iloc - 299)
end = min(len(df), start + 300)
if end - start < 300:
    start = max(0, end - 300)
df_lev = df.iloc[start:end]

distanceslevenshtein = []
for _, row in df_lev.iterrows():
    if row["title"] == "The Matrix Revolutions":
        continue
    title = row["title"]
    overview = str(row["overview"])[:150]
    distance = levenshtein_distance(Matrix_Overview, overview)
    distanceslevenshtein.append((title, distance))
distanceslevenshtein.sort(key=lambda x: (x[1], x[0]))
top10levenshtein = distanceslevenshtein[:10]
for titulo, distancia in top10levenshtein:
    print(f"{titulo}: {distancia:.3f}")
```

```python
cols = ["budget", "revenue", "popularity"]
X = df[cols + ["title"]].copy()
X[cols] = (X[cols] - np.nanmean(X[cols], axis=0)) / np.nanstd(X[cols], axis=0, ddof=0)
matrix_title = "The Matrix Revolutions"
i_matrix = df.index[df["title"] == matrix_title][0]
v_matrix = X.loc[i_matrix, cols].values
distanceseuclidiana = []
for _, row in X.iterrows():
    if row["title"] == "The Matrix Revolutions":
        continue
    v = row[cols].values
    titulo = row["title"]
    distance = np.linalg.norm(v - v_matrix)
    distanceseuclidiana.append((titulo, distance))
distanceseuclidiana.sort(key=lambda x: (x[1], x[0]))
top10euclidiana = distanceseuclidiana[:3]
for titulo, distancia in top10euclidiana:
    print(f"{titulo}: {distancia:.3f}")
```

Conclusiones:

¿Por qué los resultados pueden diferir entre Jaccard, Levenshtein y Euclidiana?

Los resultados difieren, ya que cada uno de los algoritmos varía tanto en los atributos utilizados como en la métrica aplicada, ya que cada distancia usa una definición diferente de similitud: Jaccard compara la intersección de conjuntos, evaluando coincidencias de géneros, Levenshtein mide transformaciones de texto, hallando similitud narrativa, y la distancia Euclidiana evalúa separación geométrica entre vectores numéricos asociados a variables económicas. Por esto, a pesar de tener los mismos datos, las métricas producen diferentes vecinos más cercanos



¿si fuera a recomendar películas que distancia usaría? ¿por qué?

La métrica que utilizaría para recomendar películas similares es la distancia de Levenshtein, dado que nos genera una metrica sobre la columna overview, esta desición la tomamos teniendo en cuenta lo siguiente:
 
Las variables como presupuesto, revenue y popularidad reflejan el desempeño comercial de una película, no su contenido. Dos películas pueden tener presupuestos y popularidad similar y ser completamente diferentes. Por otra parte tenemos que la distancia de Jaccard mejora, ya que toma en cuenta los generos, los que si capturan información relevante sobre la película, pero sigue siendo demasiado general ya que dos películas pueden compartir géneros y aún asi tener tramas y públicos objetivos muy diferentes. Finalmente consideramos que entre las opciones vistas la que guarda más información sobre como es la película es el overview, que habla sobre trama personajes y a nuestro parecer es una medida más representativa del contenido real.
