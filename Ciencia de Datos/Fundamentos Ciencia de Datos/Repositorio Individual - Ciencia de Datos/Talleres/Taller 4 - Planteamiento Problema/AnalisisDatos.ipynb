```python
# Contar NaN en una columna específica
# Reemplaza 'nombre_columna' por la columna que quieras
import pandas as pd
from pathlib import Path
datos = pd.read_csv("insurance_claims.csv")
col = "_c39"
print(f"NaN en {col}:", datos[col].isna().sum())
```

```python
corr = datos.corr(numeric_only=True)

threshold = 0.9
corr_pairs = (
    corr.where(~corr.eq(1.0))  # quitar diagonal
        .stack()
        .sort_values(ascending=False)
)
print("Correlaciones directas fuertes:")
print(corr_pairs[corr_pairs >= threshold].head(10))


```
