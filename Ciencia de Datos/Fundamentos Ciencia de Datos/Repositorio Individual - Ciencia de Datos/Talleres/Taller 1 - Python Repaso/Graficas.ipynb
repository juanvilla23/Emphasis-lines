```python
import numpy as np
import matplotlib.pyplot as plt





t = np.linspace(0, 10.0, 100) #Empiza en 0 y termina en 10 en la grafica, y entre 0 y 10 vas a macar 100 puntos para que la curva sea suave 

plt.figure(1)
plt.plot(t, np.sin(t), label='sin(t)')
plt.plot(t, np.cos(t), label='cos(t)')
plt.xlabel('t')
plt.ylabel('Signal(t)')
plt.ylim(-1.5, 1.5)
plt.xlim(right=12.0)
plt.legend()
plt.show()
```
