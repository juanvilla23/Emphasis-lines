```python
mis_numeros = [10, 20, 30]

def triplicar_valores(lista):
    for i in range(len(lista)):
        lista[i] = lista[i] * 3

triplicar_valores(mis_numeros)
print(mis_numeros)
```

```python
mis_numeros = [10, 20, 30]

def obtener_triplicados(lista):
    nueva_lista = [x * 3 for x in lista]
    return nueva_lista

resultados_nuevos = obtener_triplicados(mis_numeros)

print("Original:", mis_numeros)
print("Nueva:", resultados_nuevos)
```

```python
class Punto2D:
    def __init__(self, coorX, coorY):
        self.coorX = coorX
        self.coorY = coorY

    def distanciaOrigen(self):
        return (self.coorX**2 + self.coorY**2)**(1/2)

p = Punto2D(10, 20)
print(p.distanciaOrigen())
```

```python
class Punto2D:
    def __init__(self, coorX, coorY):
        self.coorX = coorX
        self.coorY = coorY

    def distanciaOrigen(self):
        return (self.coorX**2 + self.coorY**2)**(1/2)

    def distanciaPuntos(self, p2):
        return ((self.coorX - p2.coorX)**2 + (self.coorY - p2.coorY)**2)**(1/2)

p = Punto2D(10, 20)
q = Punto2D(30, 40)

print(p.distanciaOrigen())
print(p.distanciaPuntos(q))
```
