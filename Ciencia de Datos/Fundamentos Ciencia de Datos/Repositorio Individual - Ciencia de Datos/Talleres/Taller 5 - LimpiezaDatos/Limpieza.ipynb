¿Qué columnas “deberían” ser numéricas pero salen como object? (lista)
Muestra 10 valores únicos de: YEAR, RATING, VOTES, RunTime, Gross (uno por uno)

```python
datos.columns[datos.dtypes == "object"]
datos["YEAR"].unique()[:10]
```

```python
year_txt = datos["YEAR"].astype(str).str.strip().str.strip("()")
print(year_txt)
year_txt = year_txt.str.replace("–", "-", regex=False).str.replace("—", "-", regex=False)
print(year_txt)
datos["year_start"] = pd.to_numeric(year_txt.str.extract(r"(\d{4})")[0], errors="coerce")
```
