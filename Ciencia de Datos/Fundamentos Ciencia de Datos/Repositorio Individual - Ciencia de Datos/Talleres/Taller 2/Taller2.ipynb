1. Grafique 4 curvas en ventanas separadas (por ejemplo, 4 curvas
senoidales de diferente frecuencia). Recuerde crear un vector t que
represente el eje del tiempo (coordenada x). Adiciónele atributos a
cada curva (por ejemplo, título de la gráfica, nombre del eje x, nombre
del eje y, entre otros).

```python

import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 10.0, 100) 

plt.figure(1)
plt.subplot(221) # 2 filas, 2 columna, 1er subplot
plt.plot(t, np.sin(t), label='sin(t)', color="red")
plt.xlabel('t')
plt.ylabel('Signal(t)')
plt.ylim(-1.5, 1.5)
plt.xlim(right=12.0)

plt.figure(1)
plt.subplot(222)
plt.plot(t, np.cos(t), label="con(t)",color="blue")
plt.xlabel('t')
plt.ylabel('Signal(t)')
plt.ylim(-1.5, 1.5)
plt.xlim(right=12.0)


plt.figure(1)
plt.figure(1)
plt.subplot(223)
plt.plot(t, 2*np.sin(3*t + np.pi/4), label="y=Asin(ωt+ϕ)",color="green")

plt.xlabel('t')
plt.ylabel('Signal(t)')
plt.ylim(-1.5, 1.5)
plt.xlim(right=12.0)

plt.figure(1)
plt.figure(1)
plt.figure(1)
plt.subplot(224)
plt.plot(t, np.tan(t), label="tan(t)",color="purple")
plt.xlabel('t')
plt.ylabel('Signal(t)')
plt.ylim(-1.5, 1.5)
plt.xlim(right=12.0)

plt.show

```
