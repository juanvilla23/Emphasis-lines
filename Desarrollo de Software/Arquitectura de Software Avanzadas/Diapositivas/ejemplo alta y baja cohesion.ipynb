```python
# ❌ BAJA COHESIÓN — esta clase hace demasiadas cosas no relacionadas
class Utilidades:
    def calcular_prima_seguro(self, edad, riesgo): ...
    def enviar_email(self, destinatario, asunto): ...
    def formatear_fecha(self, fecha): ...
    def generar_pdf(self, contenido): ...
    def hashear_password(self, password): ...
    # No hay un hilo conductor — esto es un "God Object"


# ✅ ALTA COHESIÓN — cada clase tiene una razón de existir
class CalculadoraPrimaSeguro:
    """Todo lo que hace, gira alrededor del cálculo de primas."""
    def calcular_prima_base(self, edad: int) -> float:
        return edad * 12.5

    def aplicar_factor_riesgo(self, prima_base: float, riesgo: str) -> float:
        factores = {"bajo": 1.0, "medio": 1.3, "alto": 1.8}
        return prima_base * factores.get(riesgo, 1.0)

    def calcular_prima_final(self, edad: int, riesgo: str) -> float:
        base = self.calcular_prima_base(edad)
        return self.aplicar_factor_riesgo(base, riesgo)
```
