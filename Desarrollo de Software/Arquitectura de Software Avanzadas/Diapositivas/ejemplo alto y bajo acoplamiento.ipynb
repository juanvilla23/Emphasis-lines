```python
from abc import ABC, abstractmethod


# ❌ ALTO ACOPLAMIENTO — el orquestador conoce detalles internos
class ServicioNotificacionEmail:
    def enviar(self, destinatario: str, mensaje: str):
        print(f"[SMTP] Enviando a {destinatario}: {mensaje}")

class ProcesadorSiniestro:
    def __init__(self):
        # Instancia directa: está acoplado a la implementación concreta
        self.notificador = ServicioNotificacionEmail()

    def procesar(self, siniestro_id: str):
        # Si mañana queremos SMS o WhatsApp, hay que editar esta clase
        self.notificador.enviar("cliente@email.com", f"Siniestro {siniestro_id} procesado")


# ✅ BAJO ACOPLAMIENTO — depende de la abstracción, no de la implementación
class Notificador(ABC):
    @abstractmethod
    def enviar(self, destinatario: str, mensaje: str): ...

class NotificadorEmail(Notificador):
    def enviar(self, destinatario: str, mensaje: str):
        print(f"[SMTP] {destinatario}: {mensaje}")

class NotificadorSMS(Notificador):
    def enviar(self, destinatario: str, mensaje: str):
        print(f"[SMS] {destinatario}: {mensaje}")

class NotificadorWhatsApp(Notificador):
    def enviar(self, destinatario: str, mensaje: str):
        print(f"[WA] {destinatario}: {mensaje}")

class ProcesadorSiniestro:
    def __init__(self, notificador: Notificador):   # Inyección de dependencia
        self.notificador = notificador              # Solo conoce el contrato

    def procesar(self, siniestro_id: str):
        self.notificador.enviar("cliente@email.com", f"Siniestro {siniestro_id} procesado")

# Cambiar canal de notificación NO toca ProcesadorSiniestro
procesador = ProcesadorSiniestro(NotificadorSMS())
procesador.procesar("SIN-2024-0891")
```
